```python
import pandas as pd
```

```python
pd.set_option("mode.chained_assignment", None)
```

```python
df = pd.read_csv("Messy_Data.csv")
```

```python
df
```

```python
df.head(20)
```

```python
df = df.drop(labels=13, axis=0)
df.reset_index(drop=True, inplace=True)



```

```python
df.head(20)
```

```python
df.drop(["Unnamed: 8"], axis = 1 , inplace= True)
```

```python
df
```

```python
df['Release year'] = pd.to_datetime(df['Release year'])
df['Release Date'] = df['Release year'].dt.strftime('%m-%d-%Y')
```

```python
df.head(20)
```

```python
df['Income'] = df['Income'].replace(',','.',regex=True)
df["Income"].replace("$ 4o8.035.783", "$ 408.035.783", inplace = True)
df
```

```python
df
```

```python
df["Country"].replace("US", "USA", inplace= True)
```

```python
df["Score"] = df['Score'].replace(',','.', regex=True)
df["Score"] = df['Score'].replace('f','',regex=True)
df["Score"] = df['Score'].replace('e','',regex=True)
df["Score"] = df['Score'].replace('-','',regex=True)
df['Score'] = df['Score'].replace(':','.',regex=True)
df["Score"] = df['Score'].str.replace('9..0','9.0')
df["Score"] = df['Score'].str.replace('8..8','8.8')
df["Score"] = df['Score'].str.replace('8.7.','8.7')
df["Score"] = pd.to_numeric(df["Score"], errors= "coerce")
```

```python
df.drop(["Release year"], axis = 1 , inplace= True)
```

```python
df
```

```python
df = df[["IMBD title ID", "Original title", "Release Date", "Genre", "Duration", "Country", "Content Rating", "Director", "Income", " Votes ", "Score"]]
df
```

```python
df['Duration'] = df['Duration'].replace('Inf','')
df["Duration"] = pd.to_numeric(df["Duration"], errors= "coerce")
meanDuration = df['Duration'].mean()
df["Duration"].fillna(meanDuration, inplace= True)
```

```python
df['Content Rating'].fillna("Not Rated", inplace= True)
df
```

```python
df.to_csv('new.csv',index=False)
```
